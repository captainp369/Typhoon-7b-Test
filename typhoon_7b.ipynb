```python
import torch
from transformers import pipeline
import os
```

```python
# load model
typhoon_pipeline = pipeline(
                      model="scb10x/typhoon-7b",
                      torch_dtype=torch.bfloat16,
                      trust_remote_code=True,
                      device_map="auto"
                    )
```

```python
def get_response(input):
  typhoon_response = typhoon_pipeline(input,max_new_tokens=500)
  return typhoon_response[0]["generated_text"]
```

```python
# Let's have a conversation
prompt = "ช่วยแนะนำสถานที่ท่องเที่ยว10แห่งในจังหวัดเพชรบุรีพร้อมบอกเหตุผล"
print("Typhoon 7b response:\n"+get_response(prompt))
```
